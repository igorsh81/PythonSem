```python
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguins.loc[penguins['bill_length_mm'] > 35 & (penguins['flipper_length_mm'] <= 42), 'Floy'] = 'средний'
penguins.loc[penguins['bill_length_mm'] > 42, 'Floy'] = 'длинный'
penguins.loc[penguins['bill_length_mm'] <= 35, 'Floy'] = 'маленький'
```

```python
penguins.head(20)
```

```python
penguins.groupby('Floy')['body_mass_g'].mean().plot(kind='bar')
```
